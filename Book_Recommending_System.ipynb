##popularity based recommendation system

```python
rating_with_name=ratings.merge(books,on='ISBN')
```

```python
num_rating_df=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
# Convert 'Book-Rating' to numeric, coercing errors to NaN
rating_with_name['Book-Rating'] = pd.to_numeric(rating_with_name['Book-Rating'], errors='coerce')

# Drop rows where 'Book-Rating' is NaN
rating_with_name_cleaned = rating_with_name.dropna(subset=['Book-Rating'])

# Ensure 'Book-Rating' is of type float
rating_with_name_cleaned['Book-Rating'] = rating_with_name_cleaned['Book-Rating'].astype(float)
```

```python
# Group by 'Book-Title' and calculate the mean rating
avg_rating_df = rating_with_name_cleaned.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Display the resulting DataFrame
print(avg_rating_df)

avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
popular_df
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

##Colaborating Filtering Based Recommander System


```python
rating_with_name_cleaned
```

```python
x=rating_with_name_cleaned.groupby('User-ID').count()['Book-Rating']>200
pade_like_users=x[x].index
```

```python
filtered_rating=rating_with_name_cleaned[rating_with_name_cleaned['User-ID'].isin(pade_like_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
from typing import final
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
    # Get the index of the book in the DataFrame
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
import numpy as np

def recommend(book_name):
    # Check if the book exists in the DataFrame index
    if book_name not in pt.index:
        print(f"Book '{book_name}' not found in the dataset.")
        return

    # Get the index of the book in the DataFrame
    index = np.where(pt.index == book_name)[0][0]

    # Calculate similarity scores and get top 5 similar books
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]

    # Print the recommended books
    print(f"Recommendations for '{book_name}':")
    for i in similar_items:
        print(pt.index[i[0]])

# Example usage
recommend('The Reader')
```
